```python
import pandas as pd, numpy as np
import matplotlib.pyplot as plt, seaborn as sns

train=pd.read_csv("../data/train.csv")
test =pd.read_csv("../data/test.csv")
```

```python
missing=train.isnull().sum().sort_values(ascending=False)
missing[missing>0].head(15)
```

```python
X=train.select_dtypes(include=[np.number]).drop(columns=["SalePrice","Id"])
y=np.log1p(train["SalePrice"])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_val,y_train,y_val=train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
from sklearn.impute import SimpleImputer
imp=SimpleImputer(strategy='median')
X_train_imp=imp.fit_transform(X_train)
X_val_imp=imp.transform(X_val)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

y_train_vec = np.asarray(y_train).ravel()
y_val_vec   = np.asarray(y_val).ravel()

model = LinearRegression()
model.fit(X_train_imp, y_train_vec)

preds = model.predict(X_val_imp).ravel()

mse  = mean_squared_error(y_val_vec, preds)
rmse = np.sqrt(mse)

print("Baseline RMSE (log):", rmse)
```

```python
import os, json
os.makedirs("../reports", exist_ok=True)
with open("../reports/baseline_score.txt","w") as f:
    f.write(f"Baseline log-RMSE: {rmse:.4f}\n")
```
